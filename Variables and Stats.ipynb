## Variables:
In statistical terms are the properties with varying values of each sample. They can describe **quantities** or **qualities**


The direction of the difference or the size of difference depends on the variable if it is nominal or ordianl varibale.

### The nominal and ordinal scale:
Nominal is qualitative and Ordinal is quantitative. The Hight variable is quantitative ordinal, does not matter how we describe it in words still is a measure. 


### Interval or ratio variables measurement

The Interval or Ratio variables helps us to determine how far one measure is different from another. The difference of two individuals A - B tells how far they are while the ratio A/B tell how much it that difference. Example If the measure is Weight a = 10kg, b = 8kg then b is 2kg below, but the ratio a/b = 0/8 then a 1.25 times the other.

The Ratio variable the absence of say Weight indicates no measure.
The Interval variable, the abscence of Weight indicates that there is no difference on the individuals weight.
