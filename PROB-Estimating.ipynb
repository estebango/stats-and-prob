## Sampling with replacement.
When we replace the item that we are trying to get, the multiplication rule is just one times the other, but when we sample without replacement, we need to discount from the sample space the one item that was drawn. Exmple: getting an ace from a deck in four drawns


```python
np.random.seed(1)
def cointoss():
    if np.random.randint(0,2) == 1:
        return "HEAD"
    else:
        return "TAIL"
```

```python
outcome = 0
prob = []
for n in range(1, 10001):
    if cointoss() == "HEAD":
        outcome += 1
    prob.append(outcome/n)
```

```python
plt.plot(prob)
plt.show()
```

## Permutation:
It is an arrange of n elements taken k from them. The order matters

## Factorial:
Is is a type of permutation of N elements where we take all N elements, k = 0

```python
#Factorial
def factori(n):
    if n == 0:
        return 1
    else:
        return n*factori(n-1)
# Permutation
def permu(n,k):
    return factori(n)/factori(n-k)
```

## Combination:
Its a permutation when the order does not matter.

```python
def combi(n,k):
    return factori(n)/(factori(k)*factori(n-k))
```
